```python
#Importar libreria
import pandas as pd 

#Cargar dataser
df_calendar = pd.read_csv("calendar.csv")
df_customers = pd.read_csv("customers.csv")
df_products = pd.read_csv("products.csv")
df_sales = pd.read_csv("sales.csv")
df_stores = pd.read_csv("stores.csv")

#Ver las primeros 5 filas para revisar la info
display(df_calendar.head())

#Ver resumen tecnico (tipos de datos, conteo de filas, valores nulos)
df_calendar.info()
```

```python
#Corrección de tipo de dato de str a date
df_calendar['date'] = pd.to_datetime(df_calendar['date'])
df_calendar.info()
```

```python
display(df_customers.head())
df_customers.info()
```

```python
#Corrección de tipo de dato de str a date
df_customers['join_date'] = pd.to_datetime(df_customers['join_date'])
df_customers.info()
```

```python
display(df_products.head())
df_products.info()
```

```python
display(df_sales.head())
df_sales.info()
```

```python
#Corrección de tipo de dato de str a date
df_sales['order_date'] = pd.to_datetime(df_sales['order_date'])
df_sales.info()
```

```python
display(df_stores.head())
df_stores.info()
```

```python
#Calcular la suma total de la columna revenue 
total_ingresos = df_sales['revenue'].sum()
print(f"El total de ingresos es: ${total_ingresos}")
```

```python
suma_general = df_customers['loyalty_member'].sum()
conteo_general = df_customers['loyalty_member'].count()
print(f"De los {conteo_general} clientes, solo {suma_general} son miembros")
```

```python
pp_cacao = df_products['cocoa_percent'].mean()
print(f"El porcentaje promedio de cacao es {pp_cacao}")
```

```python
df_sales_products = pd.merge(df_products, df_sales, on='product_id', how='left')
max_ganancia = df_sales_products.groupby('product_name')['revenue'].sum().sort_values(ascending=False)
print(max_ganancia.head(1))
```

```python
df_country_sales = pd.merge(df_stores, df_sales, on='store_id', how = 'left')
cantidad_vendida = df_country_sales.groupby('country')['quantity'].sum().sort_values(ascending=False)
print(cantidad_vendida.head(1))
```

```python
df_ingresos_meses = pd.merge(df_calendar, df_sales, left_on='date', right_on='order_date', how='left')
ingresos_años = df_ingresos_meses.groupby('month')['revenue'].sum().sort_values(ascending=False)
print(ingresos_años.head(1))
```

```python
df_categoria_neta = pd.merge(df_products, df_sales, on='product_id', how='left')
ganacia_neta_max = df_categoria_neta.groupby('category')['profit'].sum().sort_values(ascending=False)
print(ganacia_neta_max.head(1))
```

```python
df_ventas_nombres = pd.merge(df_products, df_sales, on='product_id', how='left')
df_miembros_ventas = pd.merge(df_ventas_nombres, df_customers, on='customer_id', how='left')
excl_miembros = df_miembros_ventas[df_miembros_ventas['loyalty_member']==1]
excl_ventas = excl_miembros.groupby('product_name')['quantity'].sum().sort_values(ascending=False)
nombre_producto = excl_ventas.index[0]
cantidad_producto = excl_ventas.iloc[0]
print(f"El chocolate {nombre_producto} es el más comprado por los mimebros exclusivos, con una cantidad de {cantidad_producto} unidades")
```

```python
df_miembros_ventas.to_csv('tabla_ventas_para_PowerBI.csv', index=False)
print("¡Archivo listo para Power BI!")
```
